```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
positive_points=np.array([(-2,3),(0,1),(2,-1)])
negative_points=np.array([(-2,1),(0,-1),(2,-3)])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(positive_points[:,0],positive_points[:,1],color='blue',marker='o',label='Postive Class')
plt.scatter(negative_points[:,0],negative_points[:,1],color='red',marker='x',label='Negative Class')
plt.legend(loc='best')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Initial Data Points')
plt.grid(True)
plt.show()
```

```python
X=np.vstack((positive_points,negative_points))
y=np.array([1]*len(positive_points)+[-1]*len(negative_points))
```

```python

def perceptron(X, y):
    theta = np.zeros(X.shape[1]) # وزن‌ها
    theta0 = 0  # بایاس
    j=0
    for i, xi in enumerate(X):
            # بررسی شرط به‌روزرسانی وزن‌ها
        if y[i] * (np.dot(xi, theta) + theta0) <= 0:
                theta += y[i] * xi
                theta0 += y[i]
        # رسم خط تصمیم در هر مرحله
                plot_decision_boundary(theta, theta0, j + 1)
                j+=1       
    return theta, theta0
```

```python
def plot_decision_boundary(theta,theta0,iter):
    plt.figure(figsize=(8,6))
    plt.scatter(positive_points[:,0],positive_points[:,1],color='blue',marker='o',label='Postive Class')
    plt.scatter(negative_points[:,0],negative_points[:,1],color='red',marker='x',label='Negative Class')
    x_values=np.linspace(-3,3,2)
    if theta[1] !=0:
        y_values=-(theta[0]*x_values+theta0) / theta[1]
        plt.plot(x_values,y_values,label=f"Iteration{iter}")
    else:
        x_val= -theta0/theta[0]
        plt.axvline(x=x_val,label=f"Iteration{iter}")
    plt.xlim(-3,3)
    plt.ylim(-4,4)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title(f'Decision Bounarys at Iteration{iter}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
theta_final,theta0_final=perceptron(X,y)
```
